# 03 — Quantize and evaluate


```python
import json
import sys
from pathlib import Path

sys.path.insert(0, "..")
from autoquant import AutoQuantizer

root = Path("..").resolve()
quantizer = AutoQuantizer("gpt2")
quantizer.analyze_sensitivity(num_samples=50)
cfg = quantizer.create_config(0.15)

with open(root / "config.json", "w", encoding="utf-8") as f:
    json.dump(cfg["config"], f, indent=2)

quantizer.quantize(str(root / "config.json"), str(root / "quantized_gpt2"))
print(quantizer.get_report())
```

```python
import subprocess
import sys
from pathlib import Path

root = Path("..").resolve()
r = subprocess.run(
    [sys.executable, str(root / "evaluate.py"), str(root / "quantized_gpt2")],
    cwd=str(root),
    capture_output=True,
    text=True,
)
print(r.stdout)
if r.returncode != 0:
    print(r.stderr)
```
